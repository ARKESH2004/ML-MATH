```python
import pandas as pd

data = [
['Sunny','Warm','Normal','Strong','Warm','Same','Yes'],
['Sunny','Warm','High','Strong','Warm','Same','Yes'],
['Rainy','Cold','High','Strong','Warm','Change','No'],
['Sunny','Warm','High','Strong','Cool','Change','Yes'],
['Cloudy','Warm','Normal','Strong','Warm','Same','Yes']
]

columns=['Sky','AirTemp','Humidity','Wind','Water','Forecast','EnjoySport']

df=pd.DataFrame(data,columns=columns)

df.to_csv("trainingdata.csv",index=False)

print(df)
```

```python
import csv

hypo = ['%','%','%','%','%','%']

with open('trainingdata.csv') as csv_file:
    readcsv = csv.reader(csv_file, delimiter=',')
    data = []
    print("The given training examples are:\n")
    for row in readcsv:
        print(row)
        if row[-1].upper() == "YES":
            data.append(row)

print("\nPositive examples are:\n")

for x in data:
    print(x)
TotalExamples = len(data)

print("\nInitial Hypothesis:")
print(hypo)
d = len(data[0]) - 1
hypo = data[0][:d]
print("\nHypothesis after first example:")
print(hypo)
for i in range(1, TotalExamples):
    for j in range(d):
        if hypo[j] != data[i][j]:
            hypo[j] = '?'

    print("\nStep",i,"Hypothesis:")
    print(hypo)

print("\nFinal Maximally Specific Hypothesis:")
print(hypo)
```
